```python
import netCDF4 as nc
import datetime
import numpy as np
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
avhrr_file = nc.Dataset("/lustre/storeB/users/mikhaili/icearc-avhrr-sic/data/avhrr_iceconc_daily.nc")
```

```python
osisaf_file = nc.Dataset("/lustre/storeB/users/mikhaili/icearc-avhrr-sic/data/osisaf_iceconc_daily.nc")
```

```python
get_dates = lambda timestamp: datetime.datetime.fromtimestamp(timestamp) # .replace(minute=0, hour=0, second=0, microsecond=0)
```

```python
avhrr_time = avhrr_file.variables['time'][:]
avhrr_time_list = np.array(map(get_dates, avhrr_time))
```

```python
osisaf_sic = osisaf_file.variables['ice_conc'][:]
```

```python
avhrr_sic = avhrr_file.variables['ice_conc'][:]
```

```python
os_iceconc = np.ma.array(osisaf_sic.data, mask = osisaf_sic.mask + avhrr_sic.mask)
```

```python
extent_t = lambda x: np.ma.sum((x>0)*(x<=100), dtype=np.float)
tt_extent = np.ma.array(map(extent_t, os_iceconc))
```

```python
lowlim = 80
extent_o = lambda x: np.ma.sum((x>lowlim)*(x<=100), dtype=np.float)
to_extent = np.ma.array(map(extent_o, os_iceconc))
ta_extent = np.ma.array(map(extent_o, avhrr_sic))
extent_fraction = (to_extent-ta_extent)/tt_extent*100.
extent_diff = (to_extent-ta_extent)*100/100000.
```
