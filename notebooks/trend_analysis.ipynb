```python
def timestamps_to_timeseries(timestamps, start_date='19700101', freq=None):
    if freq is 'seconds':
        timedelta = pd.Series([ pd.Timedelta(seconds=i) for i in timestamps ])
    elif freq is 'days':
        timedelta = pd.Series([ pd.Timedelta(days=i) for i in timestamps ])
    else:
        raise TypeError('Frequency not understood or not provided')
    timeindex_tmp = timedelta + pd.Timestamp(start_date)
    timeindex = timeindex_tmp.apply(lambda d: d.replace(day=15, hour=12, minute=0, second=0))
    
    return timeindex

avhrr_timestamps = avhrr_file.variables['time'][:]
avhrr_timeindex = timestamps_to_timeseries(avhrr_timestamps, freq='seconds')

osisaf_timestamps = osisaf_file.variables['time'][:]
osisaf_timeindex = timestamps_to_timeseries(osisaf_timestamps, freq='seconds', start_date='19780101')
```

```python
def convert_boolean2numeric(array):
    a = np.array(array)
    m = np.flatnonzero(a)
    return m

def match_timeseries(array1, array2):
    '''
    Args:
        array1 (pandas.Series) input array
         array2 (pandas.Series) input array
    Returns:
         timemask (numpy.ndarray) boolean mask
     '''
    
    array1_timemask = []; array2_timemask = []

    for idx, i in enumerate(array1.values):
        array1_timemask.append(i in array2.values)
    
    for idx, i in enumerate(array2.values):
        array2_timemask.append(i in array1.values)
    
    mask1 = convert_boolean2numeric(array1_timemask)
    mask2 = convert_boolean2numeric(array2_timemask)
    
    return mask1, mask2
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

def make1d(array, lowlim=None, highlim=None, scale=(10**2)/1000000.):
    ndim = array.shape
    if ndim < 3:
        raise Exception('Wrong dimensions, must be 3D array')
    return np.array(map(lambda x: x[x>lowlim].compressed().mean()*x[x>lowlim].compressed().shape[0] * scale, array))

def get_fraction(avhrr, osisaf, time_index, lowlim=None):
    
    total_osisaf_area = make1d(osisaf, lowlim=0)
    avhrr_1d  = make1d(avhrr, lowlim=lowlim)
    osisaf_1d = make1d(osisaf, lowlim=lowlim)
    diff = (avhrr_1d - osisaf_1d)/total_osisaf_area*100
    df = pd.DataFrame({'diff': diff}, index=pd.DatetimeIndex(time_index)).dropna()
    s=seasonal_decompose(df, freq=5, model='additive')
    return s

AO_diff = get_fraction(avhrr_iceconc_corrected_ma, osisaf_iceconc_ma, osisaf_time[:], lowlim=15)
```

```python
from pyresample.plot import area_def2basemap
def get_quicklook(area_def, data, vmin=None, vmax=None,
                   label='Variable (units)', num_meridians=45,
                   num_parallels=10, coast_res='c', cmap='jet'):
    """Get default Basemap matplotlib plot
    """

    if area_def.shape != data.shape:
        raise ValueError('area_def shape %s does not match data shape %s' %
                         (list(area_def.shape), list(data.shape)))
    import matplotlib.pyplot as plt
    bmap = area_def2basemap(area_def, resolution=coast_res)
    bmap.drawcoastlines()
    if num_meridians > 0:
        bmap.drawmeridians(np.arange(-180, 180, num_meridians))
    if num_parallels > 0:
        bmap.drawparallels(np.arange(-90, 90, num_parallels))
    if not (np.ma.isMaskedArray(data) and data.mask.all()):
        col = bmap.imshow(data, origin='upper', vmin=vmin, vmax=vmax, cmap=cmap)
        plt.colorbar(col, shrink=0.5, pad=0.05).set_label(label)
    return plt

a= nc.Dataset('/lustre/storeB/users/mikhaili/icearc-avhrr-sic/data/sic-avhrr-gac-vis06-gacv2-10k/2011/08/mean.nc')
ai = a.variables['ice_conc'][0,:,:]
ar = a.variables['re0609'][0,:,:]

```
