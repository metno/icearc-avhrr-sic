```python
# Derive 8 days mean series of AVHRR ice concentration
def yrange(n, m):
    '''
    Args:
        n (iterable) : array to be iterated through
        m (int) : length of the output list
    Returns:
        generator
    '''
    
    i = 0
    while i < n.shape[0]:
        yield n[i:i+m]
        i += m
        
time_range = 8
time_list = avhrr_timestamps_ma        
data = avhrr_iceconc_ma

ndim = data.shape[0]

avhrr_iceconc_8days = np.ma.ones(((ndim/time_range)+1, data.shape[1], data.shape[2]))
avhrr_timestamps_8days = []

for i, j in enumerate(yrange(data,time_range)):
    avhrr_iceconc_8days[i, : , :] = np.ma.average(j, axis=0)

for k, l in enumerate(yrange(time_list, time_range)):
    avhrr_timestamps_8days.append(datetime.datetime.fromtimestamp(np.mean(l)))
    
# Mask out the data outside of the summer months
avhrr_8days_timeindex = pd.Series(avhrr_timestamps_8days)
timemask = (avhrr_8days_timeindex.dt.month>=earliest_month) * (avhrr_8days_timeindex.dt.month <=latest_month )
avhrr_iceconc_8days = avhrr_iceconc_8days[timemask.values]
avhrr_timestamps_8days = np.array(avhrr_timestamps_8days)[timemask.values]

# mask out land and climatological extent please
avhrr_iceconc_8days = np.ma.array(avhrr_iceconc_8days, mask = avhrr_iceconc_8days.mask+(avhrr_iceconc_8days==200) + (extent_mask==False))

avhrr_iceconc_8days_data = np.where(avhrr_iceconc_8days.mask==True, -999, avhrr_iceconc_8days.data)
# avhrr_iceconc_8days_data = np.where(avhrr_iceconc_8days.data == 200, -999, avhrr_iceconc_8days.data)
```

```python
# Resample AVHRR data to the MODIS grid
area_12k = pr.utils.load_area('/lustre/storeB/users/mikhaili/icearc-avhrr-sic/codeshop/compute_sic/areas.cfg', 'nsidc_stere_north_12k_45deg')
area_10k = pr.utils.load_area('/lustre/storeB/users/mikhaili/icearc-avhrr-sic/codeshop/compute_sic/areas.cfg', 'nsidc_stere_north_10k')

valid_input_index, valid_output_index, index_array, distance_array = \
                pr.kd_tree.get_neighbour_info(area_10k, area_12k, 10000)


sigma = pr.utils.fwhm2sigma(10000 * 1.5)
gauss = lambda r: np.exp(-r ** 2 / float(sigma) ** 2)

new_data = np.ones((avhrr_iceconc_ma.shape[0], area_12k.shape[0], area_12k.shape[1]))

for i in range(avhrr_iceconc_ma.shape[0]):
    new_data[i,:,:] = pr.kd_tree.get_sample_from_neighbour_info('custom', area_12k.shape,
                                                                   avhrr_iceconc_ma.data[i,:,:],
                                                                   valid_input_index,
                                                                   valid_output_index,
                                                                   index_array,
                                                                   distance_array = distance_array,
                                                                   weight_funcs = gauss,
                                                                   with_uncert=False,fill_value=-999)

avhrr_iceconc_ma_res = np.ma.array(new_data, mask=(new_data<0)+(new_data>180))
imshow(avhrr_iceconc_ma_res[-8]);colorbar()
```

```python
# Function to convert 3D data into 1D
def make1d(array, lowlim=iceconc_lowlim, highlim=iceconc_highlim, scale=(12.5**2)/1000000.):
    return np.array(map(lambda x: np.ma.sum((x>lowlim)*(x<=highlim), dtype=np.float), array))*scale


def make1d(array, lowlim=iceconc_lowlim, highlim=iceconc_highlim, scale=(12.5**2)/1000000.):
    return np.array(map(lambda x: (x[(x>=lowlim)*(x<=highlim)].compressed().shape[0])/scale, array[:]))

avhrr_iceconc_1d = make1d(np.ma.array(avhrr_iceconc_ma_res_adjusted, mask = avhrr_iceconc_ma_res_adjusted.mask * (modis_lats>max_latitude)))
modis_iceconc_1d = make1d(np.ma.array(100. - modis_owf_ma + 3, mask=(modis_lats>max_latitude) * modis_owf_ma.mask))

figure(figsize=(20,10))

plot(modis_timestamps_ma, modis_iceconc_1d, 'bd', markersize=4, label='MODIS ice extent based on ice conc')
plot(avhrr_timestamps_ma, avhrr_iceconc_1d, 'gd', label='AVHRR ice extent')
legend()
title('Extent in Mil. Sq. Km')

modis_iceconc_dataframe = pd.DataFrame(data={'MODIS SIC': modis_iceconc_1d }, index=modis_timestamps_ma)
avhrr_iceconc_dataframe = pd.DataFrame(data={'AVHRR SIC': avhrr_iceconc_1d}, index=avhrr_timestamps_ma)
iceconc_dataframe = modis_iceconc_dataframe.join(avhrr_iceconc_dataframe,how='outer')

iceconc_dataframe_mean = iceconc_dataframe.resample('8D').mean()
iceconc_dataframe_mean.plot.scatter(x='AVHRR SIC', y='MODIS SIC')
print iceconc_dataframe_mean.corr(method='pearson')
#iceconc_dataframe_mean
iceconc_dataframe_mean['AVHRR SIC'] - iceconc_dataframe_mean['MODIS SIC']
```
