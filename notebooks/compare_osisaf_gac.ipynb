```python
import netCDF4 as nc
import datetime
import numpy as np
import pandas as pd
```

```python
avhrr_file = nc.Dataset("/lustre/storeB/users/mikhaili/icearc-avhrr-sic/data/stats-vis06-gacv2-10k/avhrr-sic-monmean-merged.nc")
```

```python
osisaf_file = nc.Dataset("/lustre/storeB/users/mikhaili/icearc-avhrr-sic/data/osisaf/osisaf_merged_198205-200809.nc")
```

```python
get_dates = lambda timestamp: datetime.datetime.fromtimestamp(timestamp) # .replace(minute=0, hour=0, second=0, microsecond=0)
```

```python
avhrr_time = avhrr_file.variables['time'][:]
avhrr_time_list = np.array(map(get_dates, avhrr_time))
```

```python
epoch_offset = 252460800 # Osi-saf counts time from 1978-01-01 00:00:00
```

```python
osisaf_time = osisaf_file.variables['time'][:]
osisaf_time_list = np.array(map(get_dates, osisaf_time + epoch_offset))
```

```python
osisaf_year_list = np.array(map(lambda x: x.year, osisaf_time_list))
avhrr_year_list = map(lambda x: x.year, avhrr_time_list)
```

```python
def get_extent(iceconc, lowlim=15):
    extent = np.ma.array(iceconc, mask = iceconc.mask + (iceconc < lowlim) + (iceconc > 100))
    return extent
```

```python
lowlim = 50
years_list = range(1983,2008)

# years_list = [1983,2007]
extent_o = lambda x: np.ma.sum((x>lowlim)*(x<=100), dtype=np.float)
extent_a = lambda x: np.ma.sum((x>lowlim)*(x<=100), dtype=np.float)
total_o_extent = lambda x: np.ma.sum((x>0)*(x<=100), dtype=np.float)

years_dict = {}

for i in years_list:
        
    a_iceconc = avhrr_file.variables['ice_conc'][np.flatnonzero(osisaf_year_list==i)]
    o_iceconc = osisaf_file.variables['ice_conc'][np.flatnonzero(osisaf_year_list==i)]
    o_iceconc = np.ma.array(o_iceconc.data, mask = o_iceconc.mask + a_iceconc.mask)
    
    time = osisaf_time_list[np.flatnonzero(osisaf_year_list==i)]
    
    olist= np.array(map(extent_o, o_iceconc))
    alist = np.array(map(extent_a, a_iceconc))
    o_total_extent = np.array(map(total_o_extent, o_iceconc))
    
    # Replace year in the datetime object to trick pandas into thinking
    # we are dealing with the data from the same time range.
    # This is to obtain spread statistics
    time_new = np.array(map(lambda x: x.replace(year=1900), time))
    
    diff = (olist - alist)/o_total_extent*100
    years_dict[i]= pd.Series(data = diff, index=time_new)

    # print len(time_new), tsa.shape, tst.shape, tso.shape, diff.shape

df = pd.DataFrame(data=years_dict)

# Compute standard deviation for each day of the three-month period
std = df.std(axis=1).values
mean = df.mean(axis=1).values
```
