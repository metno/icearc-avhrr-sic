```python
import netCDF4 as nc
import datetime
import numpy as np
```

```python
avhrr_file = nc.Dataset("/lustre/storeB/users/mikhaili/icearc-avhrr-sic/data/avhrr_iceconc_daily.nc")
```

```python
osisaf_file = nc.Dataset("/lustre/storeB/users/mikhaili/icearc-avhrr-sic/data/osisaf_iceconc_daily.nc")
```

```python
get_dates = lambda timestamp: datetime.datetime.fromtimestamp(timestamp).replace(minute=0, hour=0, second=0, microsecond=0)
```

```python
avhrr_time = avhrr_file.variables['time'][:]
avhrr_time_list = np.array(map(get_dates, avhrr_time))
```

```python
epoch_offset = 252460800 # Osi-saf counts time from 1978-01-01 00:00:00
```

```python
osisaf_time = osisaf_file.variables['time'][:]
osisaf_time_list = np.array(map(get_dates, osisaf_time + epoch_offset))
```

```python
osisaf_year_list = np.array(map(lambda x: x.year, osisaf_time_list))
avhrr_year_list = map(lambda x: x.year, avhrr_time_list)
```

```python
a_iceconc = avhrr_file.variables['ice_conc'][np.flatnonzero(osisaf_year_list==2008)]
o_iceconc = osisaf_file.variables['ice_conc'][np.flatnonzero(osisaf_year_list==2008)]
time = osisaf_time_list[np.flatnonzero(osisaf_year_list==2008)]
```

```python
def get_extent(iceconc, lowlim=15):
    extent = np.ma.array(iceconc, mask = iceconc.mask + (iceconc < lowlim) + (iceconc > 100))
    return extent
```

```python
figure(figsize=(15,7))
clf()
from matplotlib.dates import YearLocator, MonthLocator, DateFormatter
from matplotlib.dates import YearLocator, MonthLocator, DateFormatter, DayLocator

years = YearLocator()   # every year
months = MonthLocator()  # every month
days = DayLocator()
dateFmt = DateFormatter('%b %d')

ax = gca()
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(dateFmt)
ax.xaxis.set_minor_locator(days)
ax.autoscale_view()

# pylab.ylim(0, 250)
# plt.yticks(np.linspace(0,250,6,endpoint=True))

lowlim = 80
years_list = [ 2000, 2004, 2007]
extent_o = lambda x: np.ma.sum((x>lowlim)*(x<=100), dtype=np.float)
extent_a = lambda x: np.ma.sum((x>lowlim)*(x<=100), dtype=np.float)
total_o_extent = lambda x: np.ma.sum((x>0)*(x<=100), dtype=np.float)

for i in years_list:
    a_iceconc = avhrr_file.variables['ice_conc'][np.flatnonzero(osisaf_year_list==i)]
    o_iceconc = osisaf_file.variables['ice_conc'][np.flatnonzero(osisaf_year_list==i)]
    o_iceconc = np.ma.array(o_iceconc.data, mask = o_iceconc.mask + a_iceconc.mask)
    time = osisaf_time_list[np.flatnonzero(osisaf_year_list==i)]
    olist= np.array(map(extent_o, o_iceconc))
    alist = np.array(map(extent_a, a_iceconc))
    o_total_extent = np.array(map(total_o_extent, o_iceconc)) 
    plot(map(lambda x: x.replace(year=1) , time), (olist-alist)*100./1000000.)
    # plot(map(lambda x: x.replace(year=1) , time), (olist-alist)/o_total_extent * 100.)
    plot(map(lambda x: x.replace(year=1) , time), o_total_extent/10000.)
    legend(years_list)
    xlabel('days')
    ylabel('% of total OSI SAF sea ice extent')
    title('Difference between OSI SAF and AVHRR based sea ice extent as percentage of the total OSI SAF extent')
```
